```python
grid_size = 5
wind = 0.1
discount = 0.9
horizon = 10
start_state = 15
feature_dim = 2
p1 = 0  # position of culvers burger joint
p2 = 24 # position of charging station

theta = 5*np.ones((2,1))


gw = BlockingGridworld(grid_size,wind,discount,horizon,start_state, feature_dim, p1,p2,theta)
```

```python
Pa1 = gw.transition_probability[:,0,:]
Pa2 = gw.transition_probability[:,1,:]
Pa3 = gw.transition_probability[:,2,:]
Pa4 = gw.transition_probability[:,3,:]

P = [Pa1,Pa2,Pa3,Pa4]
```

```python
Reach = {}

T = 7

# reach at time zero
Reach[str(0)] = [gw.start_state]

for t in range(1,T):
    
    pre = Reach[str(t-1)]
  
    pre_vec = np.zeros((1,gw.n_states))
    
    # create the vector of pre states 
    for s in pre:
        pre_vec[0][s] = 1.0
    
    # normalize
    pre_vec = pre_vec/pre_vec.sum()
    
    post_state = []
    
    for pre_state in pre:
        for Pa in P:
            transfer = pre_vec@Pa
            
            for index in np.argwhere(transfer> 0.0):
                post_state.append(index[1])
                
    post_state = list(set(post_state))
    Reach[str(t)] = post_state
```
